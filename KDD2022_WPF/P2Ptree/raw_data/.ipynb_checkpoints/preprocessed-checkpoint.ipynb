```python
def plot_abnormal(name,data,abnormal_idx):
    x1 = data.loc[abnormal_idx,["Wspd"]]
    y1 = data.loc[abnormal_idx,["Patv"]]
    x2 = data.loc[~abnormal_idx,["Wspd"]]
    y2 = data.loc[~abnormal_idx,["Patv"]]
    print("turbine",name,len(x1)/len(data))
    plt.figure(figsize=(10, 6))
    plt.scatter(x2, y2,s=5)
    plt.scatter(x1, y1,s=5)
    plt.xlabel("wind")
    plt.ylabel("power")
    plt.show()

def pyod_KNN(name,data,C = 0.05, show=False):
    WIND = "Wspd"
    clf = knn.KNN(contamination=C,n_neighbors=10)
    norm_data = data[[WIND,"Patv"]].copy()
    norm_data[WIND] = norm_data[WIND]/max(norm_data[WIND])
    norm_data["Patv"] = norm_data["Patv"]/max(norm_data["Patv"])
    clf.fit(norm_data)
    judge_point = -1

    y_label = clf.labels_
    good_idx = y_label==0
    bad_idx = y_label==1
#=================================
    good_data =  data.loc[good_idx,:]
    bad_data = data.loc[bad_idx,:]
    print("turbine:{} good_num:{} bad_num:{}".format(name,len(good_data),len(bad_data)))
    if show==True:
        plt.figure(figsize=(10, 6))
        plt.scatter(good_data[WIND], good_data["Patv"],s=5)
        plt.scatter(bad_data[WIND], bad_data["Patv"],s=5)
        plt.xlabel("wind")
        plt.ylabel("power")
        plt.show()
    return good_data,bad_data

def isoforest(name,data,C = 0.05, show=False):
    WIND = "Wspd"
    clf = IsolationForest(contamination=C)
    norm_data = data[[WIND,"Patv"]].copy()
    max_wind = max(norm_data[WIND])
    max_power = max(norm_data["Patv"])
    norm_data[WIND] = norm_data[WIND]/max(norm_data[WIND])
    norm_data["Patv"] = norm_data["Patv"]/max(norm_data["Patv"])
    clf.fit(norm_data)
    y_label = clf.predict(norm_data)
    good_idx = y_label == 1
    bad_idx = y_label == -1
#=================================
    good_data =  data.loc[good_idx,:]
    bad_data = data.loc[bad_idx,:]
    print("turbine:{} good_num:{} bad_num:{}".format(name,len(good_data),len(bad_data)))
    if show==True:
        plt.figure(figsize=(10, 6))
        plt.scatter(good_data[WIND], good_data["Patv"],s=5)
        plt.scatter(bad_data[WIND], bad_data["Patv"],s=5)
        plt.xlabel("wind")
        plt.ylabel("power")
        plt.show()
    print(" iso good_num:{} bad_num:{}".format(len(good_data), len(bad_data)))
    return good_data,bad_data,max_wind,max_power,clf

def lof(name,data,C = 0.05, show=False):
    WIND = "Wspd"
    clf = LOF(n_neighbors=2, contamination=C, novelty=True)
    norm_data = data[[WIND,"Patv"]].copy()
    max_wind = max(norm_data[WIND])
    max_power = max(norm_data["Patv"])
    norm_data[WIND] = norm_data[WIND]/max(norm_data[WIND])
    norm_data["Patv"] = norm_data["Patv"]/max(norm_data["Patv"])
    clf.fit(norm_data)
    y_label = clf.predict(norm_data)
    good_idx = y_label == 1
    bad_idx = y_label == -1
#=================================
    good_data =  data.loc[good_idx,:]
    bad_data = data.loc[bad_idx,:]
    print("turbine:{} good_num:{} bad_num:{}".format(name,len(good_data),len(bad_data)))
    if show==True:
        plt.figure(figsize=(10, 6))
        plt.scatter(good_data[WIND], good_data["Patv"],s=5)
        plt.scatter(bad_data[WIND], bad_data["Patv"],s=5)
        plt.xlabel("wind")
        plt.ylabel("power")
        plt.show()
    print(" iso good_num:{} bad_num:{}".format(len(good_data), len(bad_data)))
    return good_data,bad_data,max_wind,max_power,clf

def ocsvm(name,data,C = 0.05, show=False):
    WIND = "Wspd"
    clf = LOF(n_neighbors=2, contamination=C, novelty=True)
    norm_data = data[[WIND,"Patv"]].copy()
    max_wind = max(norm_data[WIND])
    max_power = max(norm_data["Patv"])
    norm_data[WIND] = norm_data[WIND]/max(norm_data[WIND])
    norm_data["Patv"] = norm_data["Patv"]/max(norm_data["Patv"])
    clf.fit(norm_data)
    y_label = clf.predict(norm_data)
    good_idx = y_label == 1
    bad_idx = y_label == -1
#=================================
    good_data =  data.loc[good_idx,:]
    bad_data = data.loc[bad_idx,:]
    print("turbine:{} good_num:{} bad_num:{}".format(name,len(good_data),len(bad_data)))
    if show==True:
        plt.figure(figsize=(10, 6))
        plt.scatter(good_data[WIND], good_data["Patv"],s=5)
        plt.scatter(bad_data[WIND], bad_data["Patv"],s=5)
        plt.xlabel("wind")
        plt.ylabel("power")
        plt.show()
    print(" iso good_num:{} bad_num:{}".format(len(good_data), len(bad_data)))
    return good_data,bad_data,max_wind,max_power,clf
```

```python
# 修正异常值
def pair_data(turbid,train_val_data,test_data):
    print(train_val_data.columns)
    #===================
#     x_col = list(train_val_data.columns)
#     x_col.remove('TurbID')
#     x_col.remove('Day')
#     x_col.remove('Tmstamp')
#     x_col.remove('Patv')
    #===================
    x_col = ["Wspd"]
    #===================
    x_train,x_val,y_train,y_val = train_test_split(train_val_data[x_col],train_val_data["Patv"],test_size=0.2,shuffle=True)
    gbm = LGBMRegressor(objective='regression', num_leaves=31, learning_rate=0.1, n_estimators=500, )
    gbm.fit(x_train, y_train, eval_set=[(x_val, y_val)], eval_metric='l2', early_stopping_rounds=5,verbose=0 )
    pre_Y = gbm.predict(test_data[x_col])
    joblib.dump(gbm, os.path.join(".\\repair_model","repair_model_{}.pkl".format(turbid)))
    test_data["Patv"] = pre_Y
    return test_data
    
```

```python
def is_valid_func(raw_data):
    nan_cond = pd.isna(raw_data).any(axis=1)
    invalid_cond = (raw_data['Patv'] < 0) | \
                   ((raw_data['Patv'] == 0) & (raw_data['Wspd'] > 2.5)) | \
                   ((raw_data['Pab1'] > 89) | (raw_data['Pab2'] > 89) | (raw_data['Pab3'] > 89)) | \
                   ((raw_data['Wdir'] < -180) | (raw_data['Wdir'] > 180) | (raw_data['Ndir'] < -720) |
                    (raw_data['Ndir'] > 720))
    indices = ((~nan_cond) & (~invalid_cond)).values.tolist()
    print("valid:",np.sum(indices)/len(indices))
    return indices
```

```python
WASH_DATA=True
DATA_DIR = "data_del5_lof"
model = "lof"
del_rate = 0.05
linear_param_dict = {}
is_valid_dict = {}
wind_power_max_dict = {}
for turbid,part_data in part_data_list:
    print(turbid,len(part_data)) 
    #1时间编码
    part_data=part_data.copy()
    part_data["Tmstamp"] = part_data["Tmstamp"].replace(time_dict)
    is_valid = is_valid_func(part_data)
    is_valid_dict[turbid] = is_valid
    #2修改时间格式
    #part_data["Day"]=(part_data["Day"].values-1)%7
    #3填充空值：几乎都是一缺就是缺一行的
#     idx = pd.isna(part_data).sum(1)>3
#     print(part_data.loc[idx,:])
    part_data = part_data.fillna(method="ffill")
    part_data = part_data.fillna(method="bfill")
    #4官方标注的异常值
    # (1)"patv"<0=========>=0
    idx1 = part_data["Patv"]<0
    if np.sum(part_data.loc[idx1,["Patv"]].values<-20)>0: 
        raise NameError("有负值小于-20,{}".format(part_data.loc[part_data["Patv"]<-20,["Patv"]]))
    part_data.loc[idx1,["Patv"]]=0
    idx2 = ((part_data['Patv']<0.1)&(part_data['Wspd']>2.5))
    # plot_abnormal(turbid,part_data,idx2)
    idx3_1 = part_data['Pab1'] > 89
    part_data.loc[idx3_1,["Pab1"]]=89
    idx3_2 = part_data['Pab2'] > 89
    part_data.loc[idx3_2,["Pab2"]]=89
    idx3_3 = part_data['Pab3'] > 89
    part_data.loc[idx3_3,["Pab3"]]=89
    idx4 = part_data['Wdir'] < -180
    part_data.loc[idx4,["Wdir"]]=-180
    idx5 = part_data['Wdir'] > 180
    part_data.loc[idx5,["Wdir"]]=180
    idx6 = part_data['Ndir'] < -720
    part_data.loc[idx6,["Ndir"]]=-720
    idx7 = part_data['Ndir'] > 720
    part_data.loc[idx7,["Ndir"]]=720
    # 异常检测  idx2是规则检测，可以去除掉=======================================
    not_wash_data = part_data.loc[idx2,:]
    wash_data = part_data.loc[~idx2,:]
    if WASH_DATA==True:
        if model == "KNN":
            good_data,bad_data = pyod_KNN(turbid,wash_data,C=del_rate,show=False)
        elif model=="isoforest":
            good_data,bad_data,max_wind, max_power,m = isoforest(turbid,wash_data,C=del_rate,show=True)
            wind_power_max_dict[turbid] = {"wind":max_wind,"power":max_power}
            joblib.dump(m, os.path.join(".\\detect_model","detect_model_{}.pkl".format(turbid)))
        elif model == "lof":
            good_data,bad_data,max_wind, max_power,m = lof(turbid,wash_data,C=del_rate,show=True)
            wind_power_max_dict[turbid] = {"wind":max_wind,"power":max_power}
            joblib.dump(m, os.path.join(".\\detect_model","detect_model_{}.pkl".format(turbid)))
            raise 
        else:
            raise
        #异常修复====================================================================
        # bad_data和not_wash_data都是坏的
        bad_data = bad_data.append(not_wash_data)
        #     plt.scatter(bad_data["Wspd"],bad_data["Patv"])
        #     plt.show()
    else:
        good_data = wash_data
        bad_data = not_wash_data
    bad_data = pair_data(turbid,good_data,bad_data)
    #     plt.scatter(bad_data["Wspd"],bad_data["Patv"])
    #     plt.show()
    # bad_data需要做修正
    finally_data = good_data.append(bad_data)
    assert len(finally_data)==len(part_data)
    finally_data = finally_data.sort_values(by =["Day","Tmstamp"],ascending=True)
    #print(finally_data)
    finally_data.to_csv("..\\{}\\turb{}.csv".format(DATA_DIR, turbid),index=False,header=True)
    # 线性三次方拟合==================================================
    x3=finally_data["Wspd"].values**3
    x2=finally_data["Wspd"].values**2
    x1=finally_data["Wspd"].values
    l_data = pd.DataFrame({"x3":x3,"x2":x2,"x1":x1})
    l_model = linear_model.LinearRegression()
    l_model.fit(l_data,finally_data["Patv"].values)
    K=l_model.coef_
    B=l_model.intercept_
    MAX_P = np.max(finally_data["Patv"])
    PREY = K[0]*x3+K[1]*x2+K[2]*x1+B
#     plt.scatter(x1,PREY,s=3)
#     plt.scatter(x1,finally_data["Patv"].values,s=3)
#     plt.show()
    linear_param_dict[turbid]={"K":K,"B":B,"MAX_P":MAX_P} 
```
